```python
import csv
import pickle

import keras 
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
M1 = pd.read_excel("AUD_money_supply.xlsx", sheet_name="M1")
M3 = pd.read_excel("AUD_money_supply.xlsx", sheet_name="M3")
M1_YoY = pd.read_excel("AUD_money_supply.xlsx", sheet_name="M1_YoY")
M3_YoY = pd.read_excel("AUD_money_supply.xlsx", sheet_name="M3_YoY")
```

## M1&M3

```python
X_M1 = M1["年月"]
Y_M1 = M1["數值"]

X_M3 = M3["年月"]
Y_M3 = M3["數值"]
```

```python
# set plot size
fig = plt.figure()
fig.set_figheight(9)
fig.set_figwidth(16)

# draw a plot
plt.plot(X_M1, Y_M1/1000, color='r', label="M1")
plt.plot(X_M3, Y_M3/1000, color='b', label="M3")


plt.ylabel('Ammount(K)', fontsize=24)
plt.xlabel('Date', fontsize=24)
plt.title("M1 & M3", fontsize=30, fontweight='bold')

plt.grid(True)
plt.legend(loc='upper left', fontsize=20)
plt.show()
```

## M1&M3 年增率

```python
X_M1_YoY = M1_YoY["年月"]
Y_M1_YoY = M1_YoY["數值"]

X_M3_YoY = M3_YoY["年月"]
Y_M3_YoY = M3_YoY["數值"]
```

```python
# set plot size
fig = plt.figure()
fig.set_figheight(9)
fig.set_figwidth(16)

# draw a plot
plt.plot(X_M1_YoY, Y_M1_YoY, color='r', label="M1_YoY")
plt.plot(X_M3_YoY, Y_M3_YoY, color='b', label="M3_YoY")


plt.ylabel('Rate', fontsize=24)
plt.xlabel('Date', fontsize=24)
plt.title("M1_YoY & M3_YoY", fontsize=30, fontweight='bold')

plt.grid(True)
plt.legend(loc='upper left', fontsize=20)
plt.show()
```
