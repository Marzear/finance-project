```python
from datetime import datetime

import pandas as pd
from pandas.plotting import register_matplotlib_converters
import matplotlib.pyplot as plt

register_matplotlib_converters()
```

```python
df = pd.read_csv("USA_M1.csv")
```

```python
year_month = df["年月"]
m1_list = df["數值"]
```

```python
fig = plt.figure()
fig.set_figheight(9)
fig.set_figwidth(16)

x = [datetime.strptime(str(d), '%Y/%m').date() for d in year_month]
y = [int(amount.replace(",", "")) for amount in m1_list]
plt.plot(x, y)
plt.title("USA M1", fontsize=24)
plt.xlabel("Year", fontsize=20)
plt.ylabel("Amount(Billion)", fontsize=20)
plt.show()
```
